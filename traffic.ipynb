```python
from glob import glob

img_list = glob('/home/gwon/ws/ml/yolov5/Traffic/images/*.png')
label_list = glob('/home/gwon/ws/ml/yolov5/Traffic/labels/*.txt')
```

```python
print(len(img_list))
print(len(label_list))
```

```python
from sklearn.model_selection import train_test_split
train_img_list, val_img_list = train_test_split(img_list, test_size = 0.3,
random_state = 200)
print(len(train_img_list), len(val_img_list))
```

```python
with open('/home/gwon/ws/ml/yolov5/Traffic/train.txt','w') as f:
    f.write('\n'.join(train_img_list)+'\n')
with open('/home/gwon/ws/ml/yolov5/Traffic/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list)+'\n')
```
